```python
import pandas as pd

# Load the datasets
modis_data = pd.read_csv('MODIS_pre_processed_data.csv')
precipitation_data = pd.read_csv('PrecipitationData.csv')

# Display the first few rows of each dataset
modis_head = modis_data.head()
precipitation_head = precipitation_data.head()

modis_head, precipitation_head
```

```python
# Preprocessing the MODIS dataset

# Convert the 'Date' column to datetime format
modis_data['Date'] = pd.to_datetime(modis_data['Date'])

# Extract "YYYY-MM" from the 'Date' and create a new column 'YYYY-MM'
modis_data['YYYY-MM'] = modis_data['Date'].dt.to_period('M')

# Drop the original 'Date' column and the unnamed column
modis_data = modis_data.drop(columns=['Date', modis_data.columns[0]])

# Group by 'YYYY-MM' and calculate the average for each month
modis_data_monthly_avg = modis_data.groupby('YYYY-MM').mean().reset_index()

# Show the first few rows of the preprocessed MODIS data
modis_data_monthly_avg.head()
```

```python
# Preprocessing the Precipitation Data

# Convert the 'date' column to period format "YYYY-MM" for consistency
precipitation_data['date'] = pd.to_datetime(precipitation_data['date']).dt.to_period('M')

# Check the first few rows of the preprocessed Precipitation Data
precipitation_data.head()
```

```python
# Merging the two datasets on the "YYYY-MM" date column

# Renaming the date columns for consistency
modis_data_monthly_avg = modis_data_monthly_avg.rename(columns={'YYYY-MM': 'date'})

# Merging the datasets
merged_data = pd.merge(modis_data_monthly_avg, precipitation_data, on='date', how='inner')

# Showing the first few rows of the merged dataset
merged_data.head()
```

```python
# Step 1: Prepare the MODIS Data

# Load the MODIS data again
modis_data = pd.read_csv('MODIS_pre_processed_data.csv')

# Convert the 'Date' column to datetime format
modis_data['Date'] = pd.to_datetime(modis_data['Date'])

# Extract "YYYY-MM" from the 'Date' and create a new column 'YYYY-MM'
modis_data['YYYY-MM'] = modis_data['Date'].dt.to_period('M')

# Drop the original 'Date' column and the unnamed column
modis_data = modis_data.drop(columns=['Date', modis_data.columns[0]])

# Group by 'YYYY-MM' and calculate the average for each month
modis_data_monthly_avg = modis_data.groupby('YYYY-MM').mean().reset_index()

# Filter the data to include only the months of 2019 and 2022
modis_data_filtered = modis_data_monthly_avg[
    modis_data_monthly_avg['YYYY-MM'].dt.year.isin([2019, 2022])
]

# Melt the MODIS data to long format
modis_long = pd.melt(modis_data_filtered, id_vars='YYYY-MM', var_name='place', value_name='temperature')

# Show the first few rows of the prepared MODIS data
modis_long.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'Date' column is in datetime format
m['Date'] = pd.to_datetime(m['Date'], format='%Y-%m')

# Sort the data by date
m = m.sort_values(by='Date')

# Filter data for the first ward and first year available in the data
ward_name = m['Wards'].unique()[0]
year = m['Date'].dt.year.min()
ward_data = m[(m['Wards'] == ward_name) & (m['Date'].dt.year == year)]

# Ensure we have 12 months of data
if len(ward_data) == 12:
    # Convert 'Date' column to string format for plotting
    ward_data['Date'] = ward_data['Date'].dt.strftime('%Y-%m')
    
    # Create the climate graph
    fig, ax1 = plt.subplots(figsize=(10, 5))

    # Temperature data
    color = 'tab:red'
    ax1.set_xlabel('Date (Year-Month)')
    ax1.set_ylabel('Temperature (°C)', color=color)
    ax1.plot(ward_data['Date'], ward_data['Temperature'], color=color, marker='o', label='Temperature (°C)')
    ax1.tick_params(axis='y', labelcolor=color)

    # Precipitation data
    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Precipitation (mm)', color=color)
    ax2.bar(ward_data['Date'], ward_data['Precipitation'], alpha=0.6, color=color, label='Precipitation (mm)')
    ax2.tick_params(axis='y', labelcolor=color)

    # Title and layout
    plt.title(f'Climograph for {ward_name} ({year})')
    fig.tight_layout()
    plt.xticks(rotation=45)
    plt.legend(loc='upper left', bbox_to_anchor=(0.1, 1.15), ncol=2)
    plt.show()
else:
    print(f"Insufficient data for {ward_name} in {year}. Please ensure there are 12 months of data available.")
```
